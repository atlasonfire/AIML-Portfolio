### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 

1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie id in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the ids and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#create top movie rating dataframe
topMoviesDF = dataDF.groupby('movie id').agg(avg=('rating', 'mean'), count=('rating', 'count')).reset_index()

#filter for movies with more than 100 ratings
topMoviesDF = topMoviesDF[topMoviesDF['count'] > 100]


#create dataframe with merged items and movie rating based on movie id
top25DF = pd.merge(itemDF,topMoviesDF[['movie id','avg','count']],how="outer", on='movie id').dropna().sort_values(by='avg', ascending=False)

#change count from float to int then rename avg to stars because it bothers me
top25DF['count'] = top25DF['count'].astype(int)
top25DF = top25DF.rename(columns={'avg' : 'stars'}).reset_index()

#print top 25
top25DF[['movie title', 'stars']].head(25)
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women


1. There is no need to conduct statistical tests around this. Just compare the percentages and comment on the validity of the above statements.

2. you might want ot use the .sum(), .div() function here.
3. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

#### Conclusion:



```python
#create dataframe with genre, movie id, and user id
firstDF = pd.merge(itemDF,dataDF[['movie id','user id']], how='outer', on='movie id').dropna()

#create dataframe with genre, movie id, user id, and gender
secondDF = pd.merge(firstDF,userDF[['user id', 'gender']], how='outer', on='user id').dropna()

#get percentage of movies men and women watch that are drama movies
#secondDF[['Drama','gender']].groupby(['gender']).mean().sort_values(by='Drama', ascending=False)

#second try
thirdDF = secondDF[['Drama','gender']].groupby(['gender']).agg(avg=('Drama', 'mean'), count=('Drama', 'sum')).sort_values(by='avg', ascending=False)
thirdDF.rename(columns={'avg' : '% of Total', 'count' : 'Total Number'}).reset_index()
```

Women watch a higher percentage of drama, but men have watched more total drama movies.

```python
#create a dataframe grouped by gender and contains both the mean and count of sci-fi films
thirdDF = secondDF[['Sci-Fi','gender']].groupby(['gender']).agg(avg=('Sci-Fi', 'mean'), count=('Sci-Fi', 'sum')).sort_values(by='avg', ascending=False)

#rename the columns for prettiness
thirdDF.rename(columns={'avg' : '% of Total', 'count' : 'Total Number'}).reset_index()
```

Women watch a higher percentage and more total Sci-Fi than men.

```python
#create a dataframe grouped by gender and contains both the mean and count of romance films
thirdDF = secondDF[['Romance','gender']].groupby(['gender']).agg(avg=('Romance', 'mean'), count=('Romance', 'sum')).sort_values(by='avg', ascending=False)

#rename the columns for prettiness
thirdDF.rename(columns={'avg' : '% of Total', 'count' : 'Total Number'}).reset_index()
```

While women watch a higher percentage of romance movies, men have watched more total romance movies.
